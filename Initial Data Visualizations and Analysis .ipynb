### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
sns.boxplot(y= df["ConvertedComp"]);
plt.show()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
first_quartile = df['ConvertedComp'].quantile(.25)
third_quartile = df['ConvertedComp'].quantile(.75)
IQR = third_quartile - first_quartile
IQR
```

Find out the upper and lower bounds.


```python
# your code goes here
Lower_bounds= first_quartile- IQR*1.5
Upper_bounds = third_quartile + IQR*1.5
print('Lower Bounds is:', Lower_bounds, 'and Upper Bounds is:', Upper_bounds)
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
Outliers = df['ConvertedComp'] > Upper_bounds
Outliers.value_counts()
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.
